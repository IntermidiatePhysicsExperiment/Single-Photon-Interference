```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic') 			## 이 두 줄을 
plt.rcParams['axes.unicode_minus'] = False  ## 추가해줍니다. 
```

```python
path = "../LASER/A-3/"
filename = lambda x: path + "A-3_Blocker_" + str(x) + ".csv"
POSITION_COL = "blocker_position(mm)"
VOLTAGE_COL = "voltage(V)"

def get_data(x):
  data = pd.read_csv(filename(x))
  pos = data[POSITION_COL]
  volt = data[VOLTAGE_COL]
  return pos, volt

def plot_data(x):
  pos, volt = get_data(x)
  plt.plot(pos, volt, color='red', linestyle='solid', marker='o', mfc='black', mec='black', ms=3)
  plt.xlabel("Blocker Position (mm)")
  plt.ylabel("Output Voltage (V)")
  plt.title(f"Blocker Position vs Output Voltage({str(x)}호)")
  plt.show()
```

```python
plot_data(14)
```

```python
plot_data(15)
```
